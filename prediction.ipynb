```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import load_model
import pickle
import numpy as np
import pandas as pd
```

```python
model = load_model('model.h5')


with open('scaler.pkl', 'rb') as f:
    scaler = pickle.load(f)

with open('label_encoder.pkl', 'rb') as f:
    label_encoder = pickle.load(f)

with open('one_hot_encoder.pkl', 'rb') as f:
    one_hot_encoder = pickle.load(f)
```

```python
input_data={
    'CreditScore': 600,
    'Geography': 'France',
    'Gender': 'Male',
    'Age': 40,
    'Tenure': 3,
    'Balance': 60000,
    'NumOfProducts': 2,
    'HasCrCard': 1,
    'IsActiveMember': 1,
    'EstimatedSalary': 50000
}
```

```python
geo_encoded=one_hot_encoder.transform([[input_data['Geography']]]).toarray()
geo_encoded_df=pd.DataFrame(geo_encoded, columns=one_hot_encoder.get_feature_names_out(['Geography']))
geo_encoded_df
```

```python
input_data = pd.DataFrame([input_data])
```

```python
input_data = pd.concat([input_data, geo_encoded_df], axis=1).drop(columns=['Geography'],axis=1)
```

```python
input_data
```

```python
print(label_encoder.classes_)
```

```python
input_data['Gender']=1
```

```python
input_data 
```

```python
scaler_input=   scaler.transform(input_data)
```

```python
model.predict(scaler_input)
```

```python
predictionProbability=model.predict(scaler_input)[0][0]
if(predictionProbability>0.5):
    print(f"The customer is likely to churn with a probability of {predictionProbability:.2f}")
else:
    print(f"The customer is unlikely to churn with a probability of {predictionProbability:.2f}")
```
